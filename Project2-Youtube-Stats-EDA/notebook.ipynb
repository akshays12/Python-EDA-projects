## EDA of Youtube Statistics

```python
#import dependencies

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\ragav\OneDrive\Desktop\DA\Project2-Youtube-Stats-EDA\Global YouTube Statistics.csv", encoding_errors='ignore')
data
```

```python
data.shape

pd.set_option('display.max_columns', 30)

data
```

```python
data.duplicated().sum()
```

```python
data.isna().sum()
```

```python
#removing null values
df = data.dropna()

df.isna().sum()
```

```python
df.shape
```

```python
df.reset_index(drop=True, inplace=True)
df.head(10)
```

```python
df.describe
```

```python
df.columns
```

```python
#group by category and calculate subscribers mean
pd.set_option('display.float_format', '{:.3f}'.format)
subscribers_mean = df.groupby(by='category')['subscribers'].mean()
```

```python
subscribers_mean.plot(kind='bar', x='category')
plt.ylabel('mean subscribers')
plt.title('subscribers mean')
plt.show()
```

```python
channel_type_count = df['channel_type'].value_counts()
channel_type_count
```

```python
channel_type_count.head().plot(kind='pie', autopct='%1.1f%%')
plt.title('Channel type distribution')
plt.show()
```

```python
plt.scatter(df['subscribers'], df['video views'])
plt.title('subscribers vs video views')
plt.xlabel('subscribers')
plt.ylabel('views')
plt.show()
```

```python
df.columns
```

```python
subscribers_by_country = df.groupby(by='Country')['subscribers'].sum().sort_values(ascending=False)
subscribers_by_country
```

```python
subscribers_by_country.head(10).plot(kind='bar')
plt.title('subscribers by country')
plt.xlabel('country')
plt.ylabel('subscribers')
plt.show()
```

```python
pd.set_option('display.float_format', '{:.0f}'.format)
```

```python
max_subs_country = df.head(10).groupby(by='Country')['subscribers_for_last_30_days'].max()
max_subs_country.sort_values(ascending=False)
max_subs_country
```

```python
sns.scatterplot(data=max_subs_country)
plt.show()
```
